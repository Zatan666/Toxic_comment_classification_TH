```python
import os
import re
import deepcut
```

```python
orig_path = '/content/drive/Shareddrives/wakaranai/Comments_Project'
deepcut_path = os.path.join(orig_path, 'deepcut')

orig_toxic = os.path.join(orig_path, 'toxic')
new_toxic = os.path.join(deepcut_path, 'toxic')

for i in os.listdir(new_toxic):  # delete all files in deepcut folder
    os.remove(os.path.join(new_toxic, i))

orig_files_toxic = [os.path.join(orig_toxic, i) for i in os.listdir(orig_toxic)]
new_files_toxic = [os.path.join(new_toxic, i) for i in os.listdir(orig_toxic)]
for orig, new in zip(orig_files_toxic, new_files_toxic):
    with open(orig, 'r') as f:
        text = f.read()
        text = re.sub(r'@\w+ (\w* )*', '', text)  # cleaning @tag
        cut = deepcut.tokenize(text)
        neat = ' '.join([i.strip() for i in cut if i.strip()])
    with open(new, 'w') as f:
        f.write(neat)

orig_non = os.path.join(orig_path, 'non-toxic')
new_non = os.path.join(deepcut_path, 'non-toxic')

for i in os.listdir(new_non):  # delete all files in deepcut folder
    os.remove(os.path.join(new_non, i))

orig_files_non = [os.path.join(orig_non, i) for i in os.listdir(orig_non)]
new_files_non = [os.path.join(new_non, i) for i in os.listdir(orig_non)]
for orig, new in zip(orig_files_non, new_files_non):
    with open(orig, 'r') as f:
        text = f.read()
        text = ' '.join([i for i in text.split() if not i.startswith('@')])  # cleaning @tag
        cut = deepcut.tokenize(text)
        neat = ' '.join([i.strip() for i in cut if i.strip()])
    with open(new, 'w') as f:
        f.write(neat)
```
