```python
folder = '/content/drive/Shareddrives/wakaranai/Comments_Project/unclassify/1_'
txt = 1
videoId = 'MLTqFfqyrYw'

request = youtube.commentThreads().list(
        part='snippet',
        videoId=videoId,
        order='time',
        maxResults=100
    )
response = request.execute()
while 'nextPageToken' in response:
    for i in response['items']:
        thread = i['snippet']
        comment = thread['topLevelComment']['snippet']['textOriginal']
        # print(comment)
        with open(folder+str(txt)+'.txt', 'w') as f:
            f.write(comment)
            txt += 1
        reply_count = thread['totalReplyCount']
        # print(reply_count)
        
        if reply_count != 0:
            # print('----TOPIC----')
            # print(comment)
            # print(reply_count)
            
            # get replies from a comment
            comment_id = thread['topLevelComment']['id']
            request_reply = youtube.comments().list(
                part='snippet',
                parentId=comment_id,
                maxResults=100
            )
            response_reply = request_reply.execute()

            # number of replies
            replies = response_reply['items']
            # print('^', len(replies))

            for c in replies:
                reply = c['snippet']['textOriginal']
                with open(folder+str(txt)+'.txt', 'w') as f:
                    f.write(reply)
                    txt += 1
                # print('^', reply)
            # print()
    next_page = response['nextPageToken']
    request = youtube.commentThreads().list(
        part='snippet',
        videoId=videoId,
        order='time',
        pageToken=next_page,
        maxResults=100
    )
    response = request.execute()
```
